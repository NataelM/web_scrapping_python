# Ejemplo 2

Pruebas con Selenium 2

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
def test_eight_components():
    #cargar el driver
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

    driver.get("https://www.selenium.dev/selenium/web/web-form.html")

    #obtener el titulo de la pagina
    title = driver.title
    #siempre el titulo lo podemos ver en el head
    assert title == "Web form" #en donde se ve el titulo de la pagina

    #esperar a que los elementos de la pagina carguen
    driver.implicitly_wait(6) #distinto al sleep

    #buscar los elementos por nombre donde el nombre sea "my-text"
    text_box = driver.find_element(by=By.NAME, value="my-text")
    
    #buscar el elemento de enviar en el boton enviar 
    submit_button = driver.find_element(by=By.CSS_SELECTOR, value="button")

    text_box.send_keys("Selenium")
    time.sleep(6)
    submit_button.click()

    #
    message = driver.find_element(by=By.ID, value="message")
    value = message.text
    assert value == "Received!"
    time.sleep(6)
```

```python
test_eight_components()
```

# Ejmplo 5:
sueldos de la NBA

```python
import json
import time

from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
#abrir navegador
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

#entrar a la pagina 
driver.get("https://hoopshype.com/salaries/players/")

#los 'tr' son renglones dentro de una tabla de la pagina web
# y los "td" son columnas

#extrayendo todos los links de los jugadores, esto devuelve una lista
player_links = driver.find_elements(By.XPATH,
#me traigo todas las td cuyo atributo de clase sea name y de ahí busco las referencias(enlaces) 'a'
                                    "//td[@class='name']/a"
                                    )

links = []
# all_players = []

# consiguiendo los links de todos los jugadores
for player_link in player_links: 
    links.append(player_link.get_attribute('href'))

#print(links)

for link in links[:10]: #nomas para las primeras 10 entradas
    time.sleep(5)
    driver.get(link)
    player = {}
    #nombre completo
    try:
        player_fullname = driver.find_element(By.XPATH,
                                            '//div[@class="player-fullname"]'
                                            ).text
        player['fullname'] = player_fullname
    except:
        player['fullname'] = ''
        
    #equipo al que pertenecen
    try:
        player_team = driver.find_element(By.XPATH,
                                            '//div[@class="player-team"]'
                                            ).text
        player['team'] = player_team
    except:
        player['team'] = ''
    
    #renglones de los datos de los jugadores
    try:
        rows = driver.find_elements(By.XPATH,
                                    "//span[@class='player-bio-text-line-value']"
                                    )
        player['position'] = rows[0].text
        player['born']     = rows[1].text
        player['height']   = rows[2].text
        player['weight']   = rows[3].text
        player['salary']   = rows[4].text
    except:
        player['position'] = ''
        player['born']     = ''
        player['height']   = ''
        player['weight']   = ''
        player['salary']   = ''

    json_player = json.dumps(player)
    # all_players.append(json_player)
    print(json_player)
    time.sleep(2)

    with open('nba_players.json', 'a') as json_file:
        json_file.write(json_player)
        json_file.write(',\n')

driver.quit()
```

# Ejercicio 6:

peliculas de los oscares paginas hechas con Java script

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import json
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.scrapethissite.com/pages/ajax-javascript/")

links = driver.find_elements(By.XPATH, 
                           #lista de botones con los años     
                            "//a[@class='year-link']"
                            )

all_movies = []

for link in links:
    movie = {}
    year = link.text
    link.click()
    time.sleep(5)

    rows = driver.find_elements(By.XPATH, 
                                "//tbody[@id='table-body']/tr"
                                )

    for row in rows:
        title = row.find_element(By.XPATH,
                                ".//td[1]"
                                ).text

        nominations = row.find_element(By.XPATH,
                                ".//td[2]"
                                ).text

        awards = row.find_element(By.XPATH,
                                ".//td[3]"
                                ).text

        movie['year'] = year
        movie['title'] = title
        movie['noms'] = int(nominations)
        movie['awards'] = int(awards)

        json_movie = json.dumps(movie)
        all_movies.append(json_movie)
        print(json_movie)

with open('movies.json', 'a') as json_file:
    json_file.write("[\n")
    for movie in all_movies:
        json_file.write(movie)
        json_file.write(',\n')
    json_file.write("]\n")


driver.quit()
```

# Ejemplo 8

Botón de selección multiple y paginación

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select
import time
import json
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("http://amibrevalidacion.amib.com.mx:8080/srifi/#/consultas/InformacionOperadora")
time.sleep(6)

#sacar listado de las opciones de una de las cintas del seleccionador 
operators = driver.find_elements(By.XPATH, 
                                 "//select[@id='idOperadora']/option"
                                )
del operators[0] #borrando la primera enyrada porque es la opcion todos

# se itera sobre todas las posibles opciones del selector
for operator in operators:
    #encontrar las barras donde voy a seleccionar cosas
    select_2 = Select(driver.find_element(By.XPATH, 
                                          "//select[@id='idOperadora']"
                                         )
                     )    
    select_2.select_by_visible_text(operator.text)

    button = driver.find_element(By.XPATH, 
                                 "//button[@class='btn btn-default']"
                                )
    button.click() #damos click en la opcion del seleccionador

    time.sleep(7)
    operator_dict = {}
    operator_periods = []

    rows = driver.find_elements(By.XPATH, 
                                "//table[@id='table']/tbody/tr"
                               )
    
    for row in rows:
        operator_dict['period']    = row.find_element(By.XPATH, 
                                                    ".//td[1]").text
        operator_dict['name']      = row.find_element(By.XPATH, 
                                                    ".//td[2]").text
        operator_dict['assets']    = row.find_element(By.XPATH, 
                                                    ".//td[3]").text
        operator_dict['clients']   = row.find_element(By.XPATH, 
                                                    ".//td[4]").text
        operator_dict['inv_funds'] = row.find_element(By.XPATH, 
                                                      ".//td[5]").text
        #creación de los json
        json_operator = json.dumps(operator_dict)
        operator_periods.append(json_operator)
        print(json_operator)

    with open(f'{operator.text}.json', 'a') as json_file:
        json_file.write("[\n")
        for operator in operator_periods:
            json_file.write(operator)
            json_file.write(',\n')
        json_file.write("]\n")

driver.quit()
```

# Ejemplo 9

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select
import time
import json
```

```python

def get_table_information(operator_name):
    rows = driver.find_elements(By.XPATH, "//table[@id='table']/tbody/tr")
    operator_dict = {}
    operator_periods = []
    for row in rows:
        operator_dict['period'] = row.find_element(By.XPATH, ".//td[1]").text
        operator_dict['name'] = row.find_element(By.XPATH, ".//td[2]").text
        operator_dict['active'] = row.find_element(By.XPATH, ".//td[3]").text
        operator_dict['clients'] = row.find_element(By.XPATH, ".//td[4]").text
        operator_dict['investment_funds'] = row.find_element(By.XPATH, ".//td[5]").text

        json_operator = json.dumps(operator_dict)
        operator_periods.append(json_operator)
        print(json_operator)

    with open(f'{operator_name}.json', 'a') as json_file:
        for operator in operator_periods:
            json_file.write(operator)
            json_file.write(',\n')
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("http://amibrevalidacion.amib.com.mx:8080/srifi/#/consultas/InformacionOperadora")
time.sleep(6)

#sacar listado de las opciones de una de las cintas del seleccionador 
operators = driver.find_elements(By.XPATH,
                               "//select[@id='idOperadora']/option" 
                               )
del operators[0] #borrando el operador cero

for operator in operators:
    #encontrar las barras donde voy a seleccionar cosas
    select_2 = Select(driver.find_element(By.XPATH,
                                        "//select[@id='idOperadora']"))    
    select_2.select_by_visible_text(operator.text)

    button = driver.find_element(By.XPATH,
                            '//button[@class="btn btn-default"]'
                            )
    button.click()

    time.sleep(7)

    get_table_information(operator.text)


    try:
        page_last = driver.find_element(By.XPATH, "//li[@class='page-last']/a")
        page_last2 = int(page_last.text)
    except:
        pages_count = driver.find_elements(By.XPATH, "//li[@class='page-number']")
        page_last2 = len(pages_count) + 1    

    pages = range(page_last2)

    for page in pages:
        if page == (page_last2 -1):
            link = driver.find_element(By.XPATH, "//li[@class='page-next']/a")
            link.click()
        else:
            link = driver.find_element(By.XPATH, "//li[@class='page-next']/a")            
            link.click()
            get_table_information(operator.text)

driver.close()
```
